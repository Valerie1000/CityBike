## 5.	Customize the output of your map—adjust the color of the points and add an arc that connects them. Pick a suitable color palette for each. In a Markdown section, explain what settings you changed and why.

### 
For this visualization, I customized the color scheme to clearly distinguish trip origins, destinations, and flows:

    Start Points (Green): Start stations are displayed as green points, symbolizing the origin of bike trips. Green is intuitive as a “go” or starting color.
    
    End Points (Red): End stations are displayed as red points, representing the termination of trips. Red is often associated with “stop” or ending.
    
    Arcs (White to Black): The arcs connecting start and end stations use a gradient from white (light, origin side) to black (dark, destination side). This makes the flow visually directional. On a dark base map, the white arcs stand out strongly, making trip paths clear even with overlap.
    
    Base Map (Dark): The dark basemap provides high contrast so that white arcs and colored points are more visible.
    
This color scheme highlights the flow of trips from start (green/white) to end (red/black), making both the spatial distribution and directionality intuitive to interpret.
