## 4.	Create a column with the count of trips per day and merge it with the weather dataframe.

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df_temp.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df.columns
```

```python
# Weather dataframe (assuming df has PRCP, TAVG, etc.)
df['date'] = pd.to_datetime(df['date'])

# Bike rides dataframe (from started_at)
df['date_only'] = pd.to_datetime(df['started_at']).dt.normalize()  
# normalize() sets the time to midnight, so we keep only the date part
```

```python
df_group_2 = df.groupby('date')['ride_id'].count().reset_index()
df_group_2.rename(columns={'ride_id': 'bike_rides_daily'}, inplace=True)
```

```python
# Weather should already be daily data, keep only date, PRCP, TAVG
weather_df = df[['date', 'PRCP', 'TAVG']].drop_duplicates()

# Merge daily rides with daily weather
df_daily = pd.merge(
    df_group_2, weather_df,
    left_on='date',
    right_on='date',
    how='outer',
    indicator=True
)

df_daily = df_daily.sort_values('date')
```

```python
print(df_daily.head())
print(df_daily.dtypes)
print("Unique days in daily df:", df_daily['date'].nunique())
```

```python
print(df_daily['_merge'].value_counts(dropna = False))
```

```python
# Count missing values per column
print(df_daily.isna().sum())

# Show actual rows that still have missing values
missing_rows = df_daily[df_daily.isna().any(axis=1)]
print(missing_rows)
```

### the missing values are all in date column, check if the bike_rides_daily have other than 0 value for these records

```python
# Filter rows where date is missing
missing_dates = df_daily[df_daily['date'].isna()]

# Check the ride counts for those rows
print(missing_dates[['bike_rides_daily']].value_counts())

# Or explicitly list any rows with NaT and non-zero rides
nonzero_missing = missing_dates[missing_dates['bike_rides_daily'] > 0]
print(nonzero_missing)
```

```python
# Drop nan
df_daily = df_daily.dropna(subset=['date'])
```

```python
df_daily.shape
```

```python
df_daily.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df_daily['TAVG'], df_daily['bike_rides_daily'], alpha=0.5, color="teal")

plt.title("Daily Bike Trips vs. Temperature (2022)")
plt.xlabel("Average Temperature (°F)")
plt.ylabel("Number of Trips")
plt.grid(True)
plt.show()
```

### Most rides occur between 70 and 80 deg F

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12,6))

# Plot bike rides
ax1.plot(df_daily['date'], df_daily['bike_rides_daily'], color='blue', linewidth=2, label="Bike Trips")
ax1.set_xlabel("Date")
ax1.set_ylabel("Daily Bike Trips", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Second axis for temperature
ax2 = ax1.twinx()
ax2.plot(df_daily['date'], df_daily['TAVG'], color='red', linewidth=2, label="Temperature (°F)")
ax2.set_ylabel("Average Temperature (°F)", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title
plt.title("Daily Bike Trips and Temperatures in 2022")
ax1.grid(True)

plt.savefig("bike_trips_vs_temp.png", dpi=300, bbox_inches="tight")  # PNG
plt.savefig("bike_trips_vs_temp.pdf", bbox_inches="tight")          # PDF (optional)

plt.show()
```

### I used Matplotlib to create a dual-axis line chart. The primary y-axis shows daily bike trip counts, while the secondary y-axis displays average temperatures, both plotted against the same dates on the x-axis. Using ax1 for trips and ax2 (via ax1.twinx()) for temperatures allowed clear visualization of both datasets together. Added colors, labels, and the title for clarity, and saved the figure with plt.savefig() before showing it, which is, it seems, a safer habit.

## 1.	Create a histogram of the variable “tripduration” using matplotlib.

```python
import matplotlib.pyplot as plt

# Calculate trip duration in minutes
df['tripduration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Optional: remove negative or extremely long durations if any
df_clean = df[(df['tripduration_min'] > 0) & (df['tripduration_min'] < 300)]  # for example, up to 5 hours

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(df_clean['tripduration_min'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Trip Duration (minutes)")
plt.xlabel("Trip Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Use the cleaned data
data = df_clean['tripduration_min']

# Plot histogram
plt.figure(figsize=(10,6))
count, bins, ignored = plt.hist(data, bins=50, color='skyblue', edgecolor='black', density=True, alpha=0.6)

# Add smooth density curve
density = gaussian_kde(data)
x = np.linspace(data.min(), data.max(), 1000)
plt.plot(x, density(x), color='red', linewidth=2)

plt.title("Distribution of Trip Duration (minutes)")
plt.xlabel("Trip Duration (minutes)")
plt.ylabel("Density")
plt.show()
```

## 3.	Create a figure with two axes that represent different plots: one of the plots is a bar chart with the variable “rideable_type,” and the other one is a pie chart illustrating the variable “member_casual.” Hint: you need to use the OO approach here; two axes don’t mean a dual axis. Read this article on suplots, axes, and figures for some useful information!

```python
# Prepare data
rideable_counts = df['rideable_type'].value_counts()
member_counts = df['member_casual'].value_counts()

# Create figure and axes (2 subplots side by side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# --- Bar chart for rideable_type ---
ax1.bar(rideable_counts.index, rideable_counts.values, color='skyblue', edgecolor='black')
ax1.set_title('Counts by Rideable Type')
ax1.set_xlabel('Rideable Type')
ax1.set_ylabel('Count')

# --- Pie chart for member_casual ---
ax2.pie(member_counts.values, labels=member_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral','lightskyblue'])
ax2.set_title('Distribution by Member Type')

# Adjust layout
plt.tight_layout()
plt.show()

```
